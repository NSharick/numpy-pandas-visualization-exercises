```python
import matplotlib.pyplot as plt
```

```python
##1

x = range (-50, 50)
y = [n ** 2 - n + 2 for n in x]
plt.text(0,0, 'origin', fontsize = 20, color = 'red')
plt.scatter(x, y)
```

```python
##2

import math

x = range(1, 20)
y1 = [math.sqrt(n) for n in x]
plt.plot(x, y1)
plt.title('$y = sqrt(x)$')
```

```python
x = range(1, 20)
y2 = [n ** 3 for n in x]
plt.plot(x, y2)
plt.title('$y = x ** 3$')
```

```python
x = range(1, 20)
y3 = [2 ** n for n in x]
plt.plot(x, y3)
plt.title('$y = 2 ** x$')
```

```python
x = range(1, 20)
y4 = [1 / (n + 1) for n in x]
plt.plot(x, y4)
plt.title('$y = 1/(x + 1)$')
```

```python
##3
import math

x = range(1, 20)
y1 = [math.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
y4 = [1 / (n + 1) for n in x]


plt.figure(figsize = (10, 8))
plt.suptitle('All Four Together', fontsize = 20)

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('$y = sqrt(x)$')

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title('$y = x ** 3$')

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title('$y = 2 ** x$')

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title('$y = 1/(x + 1)$')
```

```python
##4

import math

x = range(0, 5)
y1 = [math.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
y4 = [1 / (n + 1) for n in x]


plt.figure(figsize = (10, 8))
plt.title('All Four - Same Figure', fontsize = 20)

plt.plot(x, y1, label = 'y = sqrt(x)', marker = 'v')

plt.plot(x, y2, label = 'y = x ** 3', marker = 'o')

plt.plot(x, y3, label = 'y = 2 ** x', marker = '*')

plt.plot(x, y4, label = 'y = 1/(x + 1)', marker = 's')

plt.grid(True, ls = '--')
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)

plt.legend()
```
