Big O Notation

```python
import matplotlib.pyplot as plt
import math

plt.figure(figsize = (12, 8))

n = range(1,25)
y1 = [0*x +1 for x in n]
y2 = [math.log(x) for x in n]
y3 = [x for x in n]
y4 = [x * math.log(x) for x in n]
y5 = [x ** 2 for x in n]
y6 = [2 ** x for x in n]
y7 = [math.factorial(x) for x in n]
y8 = [x ** x for x in n]

plt.plot(n, y1, label = 'O(1)')
plt.plot(n, y2, label = 'O(log n)')
plt.plot(n, y3, label = 'O(n)')
plt.plot(n, y4, label = 'O(n log n)')
plt.plot(n, y5, label = 'O(n^2)')
plt.plot(n, y6, label = 'O(2^n)')
plt.plot(n, y7, label = 'O(n!)')
plt.plot(n, y8, label = 'O(n^n)')

plt.ylim(0, 100)

plt.suptitle('Big O Notation', fontsize = 20)
plt.xlabel('Elements', fontsize = 18)
plt.ylabel('Operations', fontsize = 18)

plt.legend()
```
